```python
# CREATING REGRESSION & DUMMY VARIABLES
```

```python
gendum = totalsuicide[(totalsuicide["generation"] != "Millenials") & (totalsuicide["generation"] != "Generation Z")]
```

```python
gendummies = pd.get_dummies(gendum["generation"])
gendummies.head()
```

```python
gendum[gendummies.columns] = gendummies
gendum.head()
pd.get_dummies(gendum["generation"], drop_first = True)
```

```python
X1 = gendum[["G.I. Generation", "Generation X", "Silent"]]
X1full  = sm.add_constant(X1)
X1full.head()
Y = pd.DataFrame(gendum["suicides/100k pop"])
```

```python
sm.OLS(Y, X1full).fit().summary()
# regresionn for generation vs suicide
# x variable: generation
# y varaiable: suicide
```
